```python
import pandas as pd
```

```python
file1 = "Resources/schools_complete.csv"
file2 = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(file1)
student_data = pd.read_csv(file2)
```

```python
#This created a valriable witha single integer
num_schools= len((school_data['school_name']).unique())
print(num_schools)
```

```python
num_students = len((student_data['student_name'].unique()))
print(num_students)
```

```python
student_data.head()
```

```python
# Calculate the total Budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
avg_math = student_data['math_score'].mean()
avg_red = student_data['reading_score'].mean()
```

```python
passing_students= (student_data[(student_data ['math_score'] > 70)].count()['math_score'])
pm = passing_students/float(num_students)*100
print(pm)
```

```python
passing_reading = (student_data[student_data['reading_score']>70].count()['reading_score'])
rm = passing_reading/num_students*100
print(rm)
```

```python
# Overall passing Rate
op = (rm+pm)/2 
op
```

```python
raw_results = pd.DataFrame({"Total Schools": [num_schools],
                            "Total Students" : [num_students],
                            "Total Budget":[total_budget],
                            "Avgr Math" : [avg_math],
                            "Avgr Reading" : [avg_red],
                            "% Passing Math" : [pm],
                            "% Passing reading" : [rm]})


raw_results = raw_results [["Total Schools",
                            "Total Students",
                            "Total Budget",
                            "Avgr Math",
                            "Avgr Reading" ,
                            "% Passing Math",
                            "% Passing reading"]]

raw_results['Total Students'] = raw_results['Total Students'].map("{:,}".format)
raw_results['Total Budget'] = raw_results['Total Budget'].map("${:,.2f}".format)
raw_results['Avgr Math'] = raw_results['Avgr Math'].map("{:.2f}%".format)


raw_results
```

```python
#Create a tabel that sumarizes key metrics about each school
school_names = school_data['school_name'].unique()
school_names
```

```python
# Determine the school type

school_type = school_data.set_index(['school_name'])["type"]
total_students = school_data.set_index(['school_name'])["size"]
school_budget = school_data.set_index(["school_name"])["budget"]
per_student_budget = school_budget/total_students

avg_math_score = student_data.groupby(["school_name"])["math_score"].mean()
avg_reading_score = student_data.groupby(["school_name"])["reading_score"].mean()

#Calculate the pecentage passing for each schoolfor math
passing_math = student_data[(student_data["math_score"] > 70)]
passing_math_per_school = passing_math.groupby(["school_name"])["math_score"].count()
percentage_passing_math = passing_math_per_school / total_students * 100


#Calculate the pecentage passing for each schoolfor reading
passing_reading = student_data[(student_data["reading_score"] > 70)]
passing_reading_per_school = passing_reading.groupby(["school_name"])["reading_score"].count()
percentage_passing_reading = passing_reading_per_school / total_students * 100
```

```python
raw_results_two = pd.DataFrame({"School Type" : school_type,
                            "Total Students": total_students,
                            "Total School Budget" : school_budget,
                            "Per Student Budget" : per_student_budget,
                            "Average Math Score" : avg_math_score,
                            "Average Reading Score" : avg_reading_score,
                            "% Passing Math" : percentage_passing_math,
                            "% Passing Reading" : percentage_passing_reading})


raw_results_two['Total Students'] = raw_results_two['Total Students'].map("{:,}".format)
raw_results_two['Total School Budget'] = raw_results_two['Total School Budget'].map("${:,.2f}".format)
raw_results_two['Per Student Budget'] = raw_results_two['Per Student Budget'].map("${:,.2f}".format)
raw_results_two['Average Math Score'] = raw_results_two['Average Math Score'].map("{:.2f}%".format)
raw_results_two['Average Reading Score'] = raw_results_two['Average Reading Score'].map("{:.2f}%".format)


raw_results_two
```

```python
top_schools = raw_results_two.sort_values(["% Passing Math"], ascending=False)
top_schools.head(5)
```

```python
bottom_schools = raw_results_two.sort_values(["% Passing Math"], ascending = True)
bottom_schools.head(5)
```
